```python
def chatbot():
    responses={
        "hi":"Hello How can i help you?",
        "hello":"Hi there!",
        "how are you":"I'm just a bot,but i'm doing fine!",
        "what is your name":"i'm catbot.",
        "bye":"Goodbye! Have a great day!",
        "thanks":"You're welcome!",
        "help":"I can answer simple question like greetings or farewells."
    }
    print("chatbot: hi! How can i help you?.")

    while True:
        user_input=input("you:").lower().strip()
        if user_input == "bye":
            print("catbot: Goodbye!")
            break

        

        response=responses.get(user_input,"sorry,I don't understand that.")
        print(f"chatbot:{response}")
chatbot()
```
